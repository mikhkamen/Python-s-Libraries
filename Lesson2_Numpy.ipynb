```python
import numpy as np
```

# Задание 1

```python
A = np.array([[1, 6],
              [2, 8],
              [2, 11],
              [2, 10],
              [1, 7]])
```

```python
mean_a  = np.mean(A, axis=0)
mean_a 
```

# Задание 2

```python
a_centered = A - mean_a
a_centered
```

# Задание 3

```python
a_centered_sp = np.dot(a_centered[:, 0], a_centered[:, 1])
a_centered_sp
```

# Задание 4

```python
N = A.shape[0]
A_cov = a_centered_sp / (N - 1)
A_cov
```
